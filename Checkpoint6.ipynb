```python
#Next we need to split the data into a training dataset 
import random
import csv

def loadDataset(filename, split ):
    with open(filename, 'r') as csvfile:
        lines = csv.reader(csvfile)
        dataset = list(lines)
        trainingSet=[] 
        testSet=[]      
        for x in range(len(dataset)-1):
            for y in range(4):
                dataset[x][y] = float(dataset[x][y])
            if random.random() < split:
                trainingSet.append(dataset[x])
            else:
                testSet.append(dataset[x])
    return(trainingSet , testSet)



loadDataset('iris.data.txt', 0.66)

```

```python
import math
def euclideanDistance(instance1, instance2):
    length=len(instance1)-1
    distance=0
    for i in range(length):
        distance += (instance1[i]-instance2[i])**2
    distance=math.sqrt(distance)
    return distance


```

```python
data1 = [[2, 2, 2, 'a'],[2, 2, 2, 'b']]
data2 = [4, 4, 4, 'b']

for i in range(len(data1)):
    distance = euclideanDistance(data1[i],data2)
    print('Distance: ' + repr(distance))
      

```

```python
import operator

def getNeighbors(trainingSet, testInstance, k):
    
    distances = []
    for x in range(len(trainingSet)):
        length = len(testInstance)-1
        dist = euclideanDistance(trainingSet[x],testInstance)
        distances.append((trainingSet[x], dist))
    distances.sort(key=operator.itemgetter(1)) 
    
    neighbors = []         
    for x in range(k):
        neighbors.append(distances[x][0])
        return neighbors
        

```

```python
#We can test out this function as follows:

trainSet = [[2, 2, 2, 'a'], [4, 4, 4, 'b']]
testInstance = [5, 5, 5] # Selon la definition de la fonction on va utiliser que deux variables de testInstance!!!!!

k = 1

neighbors = getNeighbors(trainSet, testInstance, k) #quand je change k rien ne change!!!!!!!!!!!!!!!!

print(neighbors)
```

```python
import operator

def getResponse(neighbors):
    
    classVotes = {}
    for x in range(len(neighbors)):
        response = neighbors[x][-1] 
        classVotes[response]=0
        
    for x in range(len(neighbors)):
        for key in classVotes:
            if key == neighbors[x][-1]:
                classVotes[key]=classVotes[key]+1
            
    
    sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)
    return sortedVotes[0][0]
 
```

```python
#We can test out this function with some test neighbors, as follows:

neighbors = [[1,1,1,'a'], [2,2,2,'a'], [1,1,1,'a'], [2,2,2,'b']]
getResponse(neighbors)
```

```python
def getAccuracy(testSet, predictions):
    correct=0
    for x in range(len(testSet)):
        if testSet[x][-1]==predictions[x]:
            correct +=1
    return (correct/float(len(testSet))) * 100.0

```

```python

testSet = [[1,1,1,'a'], [2,2,2,'a'], [3,3,3,'b']]

predictions = ['a', 'a', 'a']

accuracy = getAccuracy(testSet, predictions)

print(accuracy)
```

```python


def main(filename, split,k):
    (trainingSet , testSet)=loadDataset(filename, split)
    predictions=[]
    for i in range(len(trainingSet)):
        #euclideanDistance(trainingSet, trainingSet[i])
        neighbors=getNeighbors(trainingSet, trainingSet[i], k)
        predictions.append(getResponse(neighbors)) 
    accuracy = getAccuracy(testSet, predictions)
    return (accuracy)
```

```python
main('iris.data.txt', 0.66,3)
```

# Distance Manhattan :
la distance de Manhattan: calcule la somme des valeurs absolues des différences entre les coordonnées de deux points :
        D_m(x, y) = somme(abs(xi - yi))

```python
import math
def ManhattanDistance(instance1, instance2):
    length=len(instance1)-1
    distance=0
    for i in range(length):
        distance += abs(instance1[i]-instance2[i])
    
    return distance
```

```python
import operator

def getNeighbors2(trainingSet, testInstance, k):
    
    distances = []
    for x in range(len(trainingSet)):
        length = len(testInstance)-1
        dist = ManhattanDistance(trainingSet[x],testInstance)
        distances.append((trainingSet[x], dist))
    distances.sort(key=operator.itemgetter(1)) 
    
    neighbors = []         
    for x in range(k):
        neighbors.append(distances[x][0])
        return neighbors
        

```
