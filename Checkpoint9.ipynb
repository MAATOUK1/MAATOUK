```python
import mlxtend
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
import math

data=pd.read_csv('Market_Basket_Optimisation.csv', header=None)

data.shape
```

```python
Articles=[]
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        if (pd.isnull(data.values[i,j]) == False):
            Articles.append(data.values[i,j])
Article=np.array(Articles)

df=pd.DataFrame(Article, columns=['item'])
df['article_count']=1
df_table=df.groupby('item').sum().sort_values('article_count', ascending=False).reset_index()
df_table.head(15)
```

```python
transactions=[]

for i in range(data.shape[0]):
    transactions.append([data.values[i,j] for j in range(data.shape[1]) if (pd.isnull(data.values[i,j]) == False)])
```

```python
import mlxtend
from mlxtend.preprocessing import TransactionEncoder

te=TransactionEncoder()
te_ary=te.fit(transactions).transform(transactions)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets=apriori(df, min_support=0.02, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules 
rules = association_rules(frequent_itemsets,metric="confidence",min_threshold=0.3) # associate itemsets with confidence over 70%.
rules
```

```python
from mlxtend.frequent_patterns import association_rules 
rules = association_rules(frequent_itemsets,metric="lift",min_threshold=1.01)
rules
```
